```r
media_A <- 32
media_B <- 28.4
sd_A <- 5.3
sd_B <- 5.7
n_A <- 20
n_B <- 25

perform_test <- function(alpha, media_A, media_B, sd_A, sd_B, n_A, n_B) {
  # Calcolo la deviazione standard combinata
  new_sd <- sqrt(((n_A - 1) * sd_A^2 + (n_B - 1) * sd_B^2) / (n_A + n_B - 2))
  
  # Calcolo il t-value
  t_value <- (media_A - media_B) / (new_sd * sqrt(1/n_A + 1/n_B))
  
  # Calcolo i gradi di liberta'
  df <- n_A + n_B - 2
  
  # Calcolo il p-value
  p_value <- 2 * pt(-abs(t_value), df)
  
  # Calcolo la potenza del test
  power_unilaterale <- power.t.test(n = n_A, delta = media_A - media_B, sd = new_sd, sig.level = alpha, type = "one.sample", alternative = "one.sided")$power
  power_bilaterale <- power.t.test(n = n_A, delta = media_A - media_B, sd = new_sd, sig.level = alpha, type = "one.sample", alternative = "two.sided")$power
  
  cat("Alpha =", alpha, ": t-value =", t_value, ", p-value =", p_value, "\n")
  cat("Potenza (Unilaterale):", power_unilaterale, "\n")
  cat("Potenza (Bilaterale):", power_bilaterale, "\n")
  
  par(bg="white")
  plot(c(alpha), c(power_bilaterale), type="b", pch=19, ylim=c(0,1), xlab="Alpha", ylab="Potenza", main="Potenza del test")
  points(c(alpha), c(power_unilaterale), type="b", pch=19, col="red")
  legend("bottomright", legend=c("Bilaterale", "Unilaterale"), col=c("black", "red"), pch=19)
}

livelli_alpha <- seq(0.01, 0.10, by = 0.01)
sapply(livelli_alpha, function(alpha) {
  perform_test(alpha, media_A, media_B, sd_A, sd_B, n_A, n_B)
})
```

```r
# Define the estimator function
estimate_T_star <- function(sample) {
  Q1 <- quantile(sample, probs=0.25)
  Q2 <- median(sample)
  Q3 <- quantile(sample, probs=0.75)
  T_star <- 0.25 * Q1 + 0.50 * Q2 + 0.25 * Q3
  return(T_star)
}

# Function to simulate sampling and calculate properties of the estimator
simulate_properties <- function(population_mean, population_sd, sample_size, num_simulations=10000) {
  bias_accum <- 0
  mse_accum <- 0
  
  for (i in 1:num_simulations) {
    sample <- rnorm(sample_size, mean = population_mean, sd = population_sd)
    T_star <- estimate_T_star(sample)
    
    # Calculate bias and mean squared error
    bias_accum <- bias_accum + (T_star - population_mean)
    mse_accum <- mse_accum + (T_star - population_mean)^2
  }
  
  bias <- bias_accum / num_simulations
  mse <- mse_accum / num_simulations
  
  cat("For sample size", sample_size, ":\n")
  cat("Bias of the estimator T^*:", bias, "\n")
  cat("Mean Squared Error of the estimator T^*:", mse, "\n")
  cat("\n")
}

# Check properties for different sample sizes
set.seed(123) # For reproducibility
population_mean <- 100 # Hypothetical population mean
population_sd <- 15 # Hypothetical population standard deviation

# Large sample size
simulate_properties(population_mean, population_sd, sample_size = 1000)

# Small sample size
simulate_properties(population_mean, population_sd, sample_size = 30)
```

```r
# Definire una funzione in R che determina tutti i possibili modi in cui si puo' ottenere una somma di 100 centesimi
# utilizzando le monete da 1, 2, 10, 20 e 50 centesimi.
combinazioni <- function(total, monete) {
  find_combinations <- function(total, monete, partial) {

    if (total == 0) {
      return(list(partial))
    }

    if (total < 0) {
      return(NULL)
    }
    combinations <- list()
    for (moneta in monete) {
      new_total <- total - moneta
      new_partial <- c(partial, moneta)
      more_combinations <- find_combinations(new_total, monete, new_partial)
      if (!is.null(more_combinations)) {
        combinations <- c(combinations, more_combinations)
      }
    }
    return(combinations)
  }
  
  return(find_combinations(total, monete, c()))
}

monete <- c(1, 2, 10, 20, 50)

totale_combinazioni <- combinazioni(100, monete)

print(length(totale_combinazioni))
print(totale_combinazioni)
```

```r
# Definire una funzione in R che determina tutti i possibili modi in cui si puo' ottenere una somma di 100 centesimi
# utilizzando le monete da 1, 2, 10, 20 e 50 centesimi.

combinazioni_possibili <- function(total, monete) {
  find_combinations <- function(total, monete, partial) {
    
    if (total == 0) {
      return(list(partial))
    }
    
    if (total < 0) {
      return(NULL)
    }
    combinations <- list()
    for (moneta in monete) {
      new_total <- total - moneta
      new_partial <- c(partial, moneta)
      more_combinations <- find_combinations(new_total, monete, new_partial)
      if (!is.null(more_combinations)) {
        combinations <- c(combinations, more_combinations)
      }
    }
    return(combinations)
  }
  
  return(find_combinations(total, monete, c()))
}

monete <- c(1, 2, 10, 20, 50)

totale_combinazioni <- combinazioni_possibili(100, monete)

print(length(totale_combinazioni))
print(totale_combinazioni)
```

```r
find_combinations <- function(monete, totale) {
  get_combinations <- function(monete, totale, current_combination) {
    if (totale == 0) {
      return(list(current_combination))
    }
    if (totale < 0) {
      return(NULL)
    }
    if (length(monete) == 0) {
      return(NULL)
    }
    combinations <- c(
      get_combinations(monete, totale - monete[1], c(current_combination, monete[1])),
      get_combinations(monete[-1], totale, current_combination)
    )
    combinations <- Filter(Negate(is.null), combinations)
    return(combinations)
  }
  return(get_combinations(monete, totale, c()))
}

monete <- c(50, 20, 10, 2, 1)
totale <- 100

tutte_le_combinazioni <- find_combinations(monete, totale)

print(length(tutte_le_combinazioni))
print(tutte_le_combinazioni[1:5])
```
