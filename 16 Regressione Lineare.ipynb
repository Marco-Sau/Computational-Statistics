## <span style="color:#e6550d">Retta di Regressione (Blocco 2)</span>

```r
# ? lm
lm.st <- lm(Sales ~ TV)
par(bg="white")
plot(TV, Sales, pch=16, col="blue")

abline(lm.st, col="red", lwd=4)

# Un segmento
segments(TV[1], Sales[1], TV[1], lm.st$fitted.values[1],
    col="orange", lty="dotted",lwd=3)
```

```r
par(bg="white")
plot(TV, Sales, pch=16, col="blue")

abline(lm.st, col="red", lwd=4)

# Due segmenti
segments(TV[1], Sales[1], TV[1], lm.st$fitted.values[1],
    col="orange", lty="dotted", lwd=3)
segments(TV[2], Sales[2], TV[2], lm.st$fitted.values[2],
    col="orange", lty="dotted", lwd=3)
```

```r
par(bg="white")
plot(TV, Sales, pch=16, col="blue")

abline(lm.st, col="red", lwd=4)

# Due segmenti
segments(TV[1], Sales[1], TV[1], lm.st$fitted.values[1],
    col="orange", lty="dotted", lwd=3)
segments(TV[2], Sales[2], TV[2], lm.st$fitted.values[2],
    col="orange", lty="dotted", lwd=3)
segments(TV, Sales, TV, lm.st$fitted.values,
    lty="dotted", col="orange", lwd=3)
```

## <span style="color:#e6550d">Retta Stimata (Blocco 3)</span>

#### <span style="color:#3182bd">Predizioni con 10 osservazioni</span>

```r
## Predictions
##  provare prima con 15 e poi con 1500

set.seed(123)
x <- rnorm(10)
y <- 3 + (x * 2) + rnorm(10) # y = 3.16 + 2.63 * x
lm(y ~ x) -> lll
summary(lll)

par(bg="white")
par(mfrow=c(1,1))
plot(x,y,pch=16,col="blue", main = "n = 10")
abline(lll,col="green", lty=2, lwd=4)
abline(a=3,b=2,col="red", lty=1, lwd=4)
legend("topleft", legend = c("Retta popolazione","Retta stimata"), 
       col = c("red","green"),
       lwd = 4, cex=.5)
```

#### <span style="color:#3182bd">Predizioni con 50 osservazioni</span>

```r
x <- rnorm(50)
y <- 3+x*2 + rnorm(50) # y = 3.02 + 2.05 * x
lm(y ~ x)->lll2
summary(lll2)

par(bg="white")
plot(x,y,pch=16,col="blue", main = "n = 50")
abline(lll2,col="green", lty=2, lwd=4)
abline(a=3,b=2,col="red", lty=1, lwd=4)
legend("topleft", legend = c("Retta popolazione","Retta stimata"), 
       col = c("red","green"),
       lwd = 4, cex=.5)
```

#### <span style="color:#3182bd">Predizioni con 1000 osservazioni</span>

```r
x <- rnorm(1000)
y <- 3+x*2 + rnorm(1000) # y = 3.02 + 2.05 * x
lm(y ~ x)->lll3
summary(lll3)

par(bg="white")
plot(x,y,pch=16,col="blue", main = "n = 1000")
abline(lll3,col="green", lty=2, lwd=4)
abline(a=3,b=2,col="red", lty=1, lwd=4)
legend("topleft", legend = c("Retta popolazione","Retta stimata"), 
       col = c("red","green"),
       lwd = 4, cex=.5)
```

#### <span style="color:#3182bd">Intervalli di Confidenza e Predizione</span>

```r
par(mfrow=c(1,1))
x <- rnorm(10)
y <- 3 + (x * 2) + rnorm(10) # y = 3.02 + 2.05 * x

predict(lm(y ~ x)) # restituisce i valori stimati
```

```r
# creiamo un nuovo dataset di tutti gli oggetti di cui vogliamo fare la predizione
new <- data.frame(x = seq(-3, 3, 0.05))
predict(lll, new, se.fit = TRUE) 
# lll  il modello
# new  il dataset
# se.fit = TRUE restituisce anche la deviazione standard
```

```r
pred.w.plim <- predict(lll, new, interval = "prediction")
pred.w.clim <- predict(lll, new, interval = "confidence")
par(bg="white")
matplot(new$x, cbind(pred.w.clim, pred.w.plim[,-1]),
        lty = c(1,2,2,3,3), col=c(1,2,2,3,3),type = "l", ylab = "predicted y")
legend("topleft", legend = c("Retta stimata","Intervallo di confidenza","Intervallo di predizione"),
        col = c(1,2,3), lty = c(1,2,3), lwd = 4, cex=.5)

# new$x  il dataset su cui vogliamo fare la predizione
# cbind unisce i due dataset
# pred.w.clim  il dataset con gli intervalli di confidenza
# pred.w.plim  il dataset con gli intervalli di predizione
```

## <span style="color:#e6550d">Regressione Lineare Multipla (Blocco 7)</span>

#### <span style="color:#3182bd">Regressione lineare multipla e selezione delle variabili</span>

```r
# Regressione lineare multipla e selezione delle variabili

lm.fit=lm(prezzo ~ poverta + eta, data=ds2)
summary(lm.fit)
```

```r
lm.fit=lm(prezzo ~ ., data=ds2) # modello completo
summary(lm.fit)
```

#### <span style="color:#3182bd">Stepwise</span>

```r
# ? step 
step.fit <- step(lm.fit, direction="both", trace=TRUE) 
# direction="both" = selezione mista
```

```r
summary(step.fit)
```

```r
lm.fit0 <- lm(prezzo ~ 1, data=ds2)
step.fit2 <- step(lm.fit0, scope = formula(lm.fit),
    direction="both", trace=TRUE)
```

```r
summary(step.fit2)
```

```r
par(bg="white")
pairs(ds2)
```

#### <span style="color:#3182bd">Best Subset Selection</span>

```r
# Best subset selection
library(leaps) ##caricamento package leaps
mc <- lm(prezzo ~ ., ds2) # modello completo
```

```r
y <- ds2$prezzo
x <- model.matrix(mc)[, -1]

bss.mc <- leaps(x, y, method = "Cp")
bss.mc
```

```r
which.min(bss.mc$Cp)

bss.mc$which[which.min(bss.mc$Cp),] 
# le variabili dalla 10 in poi vengono indicate con A, B, C....

bss.mc$label
```

```r
#install.packages("faraway")
library(faraway)
par(bg="white")
Cpplot(bss.mc)

names(ds2) 
# il modello selezionato da BSS non include le variabili "industriale"
# ed "eta"

# NB Stesso risultato di Stepwise selection
```
