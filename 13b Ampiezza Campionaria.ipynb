## <span style="color:#e6550d">Per la <b><i>MEDIA</i></b></span>

![Screenshot 2024-01-04 at 08.47.27.png](<attachment:Screenshot 2024-01-04 at 08.47.27.png>)

```r
sd <- 45
me <- 5 # Margine di errore
alpha <- 0.1

# Calcolo z
z <- qnorm(1 - alpha/2)
cat("Valore critico:", z, "\n")

# Calcolo la numerosit minima n
n <- (z * sd / me)^2
cat("La numerosit minima :", ceiling(n), "\n")
```

### <span style="color:#31a354">Funzione per il calcolo della numerosita' minima per la media</span>

```r
# Funzione per il calcolo della numerosit minima
num_min_media <- function(sd, me, alpha) {
    z <- qnorm(1 - alpha/2)
    n <- (z * sd / me)^2
    return(ceiling(n))
}

# Calcolo la numerosit minima n
n <- num_min_media(sd, me, alpha)
cat("La numerosit minima :", n, "\n")
```

## <span style="color:#e6550d">Per la <b><i>PROPORZIONE</i></b></span>

![Screenshot 2024-01-04 at 08.53.49.png](<attachment:Screenshot 2024-01-04 at 08.53.49.png>)

```r
me <- 0.03
alpha <- 0.05
stima <- 0.25

# Calcolo z
z <- qnorm(1 - alpha/2)
cat("Valore critico:", z, "\n")

# Calcolo la numerosit minima n
n <- (stima * (z^2)) / (me^2)
cat("La numerosit minima :", ceiling(n), "\n")
```

### <span style="color:#31a354">Funzione per il calcolo della numerosita' minima per la media</span>

```r
# Funzione per il calcolo della numerosit minima
num_min_proporzione <- function(stima, me, alpha) {
    z <- qnorm(1 - alpha/2)
    n <- (stima * (z^2)) / (me^2)
    return(ceiling(n))
}

# Calcolo la numerosit minima n
n <- num_min_proporzione(stima, me, alpha)
cat("La numerosit minima :", n, "\n")
```
